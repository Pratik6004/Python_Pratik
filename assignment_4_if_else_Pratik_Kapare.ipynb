### 40. Check if a number is Armstrong (3-digit).

```python
num=94745
digits=0
numcopy=num

sum=0

while num>0:
    num= num//10
    digits=digits+1
   
numcopy2=numcopy

while numcopy>0:
    last_digit=numcopy%10
    cube=last_digit**digits
    sum=sum+cube
    numcopy=numcopy//10

if numcopy2==sum:
    print("armstrong number")
else:
    print("not armstrong number")
```
